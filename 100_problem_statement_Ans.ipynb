```python
# Module 1: Variables, Input, Output

# 1. Write a program to print "Hello, World!".
print("Hello World!")

# 2. Take a user's name as input and print a greeting message.
name =input("enter your name : ")
print(name)

# 3. Input two integers and print their sum.
x = int(input("enter a number : "))
y = int(input("enter a number : "))
sum =  x+y
print(sum)

# 4. Input two numbers and print their product
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
print("Product:", a * b)


# 5. Convert temperature from Celsius to Fahrenheit.
c = float("enter a temp in celsius : ")
f = (c*9/5)+32
print("temp in farhrenhiet  : ", f)



# 6. Swap two variables without using a third variable
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a, b = b, a
print("After swapping:")
print("a =", a)
print("b =", b)


# 7. Calculate the area of a circle given the radius
import math
r = float(input("Enter radius: "))
area = math.pi * r * r
print("Area of circle:", area)


# 8. Input a three-digit number and print the sum of its digits
num = int(input("Enter a three digit number: "))
sum_digits = (num // 100) + ((num // 10) % 10) + (num % 10)
print("Sum of digits:", sum_digits)


# 9. Convert minutes into hours and remaining minutes
minutes = int(input("Enter minutes: "))
hours = minutes // 60
remaining = minutes % 60
print("Hours:", hours)
print("Remaining minutes:", remaining)


# 10. Calculate simple interest (P, R, T given)
P = float(input("Enter Principal: "))
R = float(input("Enter Rate: "))
T = float(input("Enter Time: "))

SI = (P * R * T) / 100
print("Simple Interest:", SI)


# 11. Convert kilometers to miles
km = float(input("Enter kilometers: "))
miles = km * 0.621371
print("Miles:", miles)


# 12. Print ASCII value of a character
ch = input("Enter a character: ")
print("ASCII value:", ord(ch))


# 13. Convert a number into binary format
num = int(input("Enter a number: "))
print("Binary:", bin(num))


# 14. Input a float and round it to two decimal places
num = float(input("Enter a float number: "))
print("Rounded value:", round(num, 2))


# 15. Find the square and cube of a number
num = int(input("Enter a number: "))
print("Square:", num**2)
print("Cube:", num**3)





# Module 2: Conditional Statements

# 16. Check whether a number is even or odd.
x =  int(input("enter a number : "))
if(x%2 == 0):
    print("x is even number : " , x)
else:
    print("x is odd number : "  , x)

# 17. Check whether a number is positive, negative, or zero.
x = int(input("entaer a number : "))
if(x>0):
  print("x is positive", x)
elif(x<0)
  print("x is neagative", x)
else:
  print("x is zero")

# 18. Find the largest among three numbers.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if(a>b and a>c):
  print("a is largest", a)
elif(b>a and b>c):
  print("b is largest ", b)
else:
  print("c is largest", c)


# 19. Check if a year is a leap year.
year = int(input("Enter year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")



# 20. Determine whether a character is a vowel or consonant.
ch = input("enter a character : ").lower()
if ch in "aeiou":
  print("character is vowel", ch)
else:
  print("character is consonant", ch)


# 21. Check if a number is divisible by both 3 and 5.
x =  int(input("enter a number : "))
if(x%3 and x%5 == 0):
    print("x is divisible by 3 & 5")
else:
    print("x is not divisible by  3 & 5")

# 22. Grade calculator (marks to grade conversion).
marks = int(input("Enter marks: "))
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
elif marks >= 40:
    print("Grade D")
else:
    print("Fail")


# 23. Check if a number is a palindrome.
x = int(input("enter a number :"))
if(x == x[::-1]):
  print("x is palindrome", x)
else:
  print("x is not plaindrome")


# 24. Check whether a triangle is valid (sum of angles).
a = int(input("Enter angle 1: "))
b = int(input("Enter angle 2: "))
c = int(input("Enter angle 3: "))

if a + b + c == 180:
    print("Valid Triangle")
else:
    print("Invalid Triangle")


# 25. Identify type of triangle (equilateral, isosceles, scalene).
a = int(input("Enter side 1: "))
b = int(input("Enter side 2: "))
c = int(input("Enter side 3: "))

if a == b == c:
    print("Equilateral Triangle")
elif a == b or b == c or a == c:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")


# 26. Check if a number lies within a given range.
x =  int(input("enter a number  : "))
if(10<=x<=50):
  print("number within a range")
else:
  print("numbe is not in a range")


# 27. Determine whether a number is prime.
num = int(input("Enter number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")


# 28. Create a simple calculator using if-elif.
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    print("Result:", a + b)
elif op == "-":
    print("Result:", a - b)
elif op == "*":
    print("Result:", a * b)
elif op == "/":
    print("Result:", a / b)
else:
    print("Invalid operator")

# 29. Check eligibility to vote (age ≥ 18).
age =  int(input("enter a age : "))
if(age>= 18):
  print("person can vote ")
else:
  print("person can not eligible to vote")


# 30. Find the second largest of three numbers.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

nums = [a, b, c]
nums.sort()
print("Second largest:", nums[1])



# Module 3: Loops

# 31. Print numbers from 1 to 100
for i in range(1, 101):
    print(i)


# 32. Print all even numbers from 1 to 50
for i in range(2, 51, 2):
    print(i)


# 33. Print multiplication table of a given number
n = int(input("Enter a number: "))
for i in range(1, 11):
    print(n, "x", i, "=", n * i)


# 34. Calculate factorial using loop
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial:", fact)


# 35. Reverse a number using loop
num = int(input("Enter a number: "))
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
print("Reversed number:", rev)


# 36. Count digits in a number
num = int(input("Enter number: "))
count = 0
while num > 0:
    num //= 10
    count += 1
print("Number of digits:", count)


# 37. Print Fibonacci series up to n terms
n = int(input("Enter number of terms: "))
a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b


# 38. Print all prime numbers up to 100
for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num)


# 39. Find sum of first n natural numbers
n = int(input("Enter n: "))
sum = 0
for i in range(1, n + 1):
    sum += i
print("Sum:", sum)


# 40. Print star pattern (right triangle)
n = 5
for i in range(1, n + 1):
    print("*" * i)


# 41. Print inverted star pattern
n = 5
for i in range(n, 0, -1):
    print("*" * i)


# 42. Check Armstrong number
num = int(input("Enter number: "))
temp = num
sum = 0
while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not Armstrong")


# 43. Find LCM of two numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
max_num = max(a, b)

while True:
    if max_num % a == 0 and max_num % b == 0:
        print("LCM:", max_num)
        break
    max_num += 1


# 44. Find GCD of two numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
while b:
    a, b = b, a % b
print("GCD:", a)


# 45. Sum of digits until single digit remains
num = int(input("Enter number: "))
while num > 9:
    sum = 0
    while num > 0:
        sum += num % 10
        num //= 10
    num = sum
print("Single digit sum:", num)




# Module 4: Strings

# 46. Count number of vowels in a string
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for ch in s:
    if ch in vowels:
        count += 1
print("Number of vowels:", count)


# 47. Reverse a string
s = input("Enter a string: ")
print("Reversed string:", s[::-1])


# 48. Check if string is palindrome
s = input("Enter a string: ")
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")


# 49. Count occurrences of a character
s = input("Enter a string: ")
ch = input("Enter character to count: ")
print("Occurrences:", s.count(ch))


# 50. Replace all spaces with hyphens
s = input("Enter a string: ")
print("Modified string:", s.replace(" ", "-"))


# 51. Convert string to uppercase without using built-in method
s = input("Enter a string: ")
result = ""
for ch in s:
    if 'a' <= ch <= 'z':
        result += chr(ord(ch) - 32)
    else:
        result += ch
print("Uppercase string:", result)


# 52. Remove duplicate characters from string
s = input("Enter a string: ")
result = ""
for ch in s:
    if ch not in result:
        result += ch
print("String without duplicates:", result)


# 53. Find longest word in a sentence
s = input("Enter a sentence: ")
words = s.split()
longest = max(words, key=len)
print("Longest word:", longest)


# 54. Check if two strings are anagrams
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not Anagrams")


# 55. Capitalize first letter of each word
s = input("Enter a sentence: ")
words = s.split()
result = " ".join(word[0].upper() + word[1:] for word in words)
print("Capitalized:", result)


# 56. Remove punctuation from a string
import string
s = input("Enter a string: ")
result = ""
for ch in s:
    if ch not in string.punctuation:
        result += ch
print("Without punctuation:", result)


# 57. Find frequency of each character
s = input("Enter a string: ")
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
print("Character frequency:", freq)


# 58. Check whether a substring exists
s = input("Enter main string: ")
sub = input("Enter substring: ")
if sub in s:
    print("Substring found")
else:
    print("Substring not found")


# 59. Split string into words and count them
s = input("Enter a sentence: ")
words = s.split()
print("Words:", words)
print("Total words:", len(words))


# 60. Format string using f-strings
name = input("Enter name: ")
age = int(input("Enter age: "))
print(f"My name is {name} and I am {age} years old.")




# Module 5: Lists

# 61. Create a list of 10 numbers and find the maximum
nums = [10, 25, 7, 45, 32, 18, 50, 3, 29, 41]
print("Maximum:", max(nums))


# 62. Find second largest element in list
nums = [10, 25, 7, 45, 32]
nums.sort()
print("Second largest:", nums[-2])


# 63. Remove duplicates from list
lst = [1, 2, 2, 3, 4, 4, 5]
unique = list(set(lst))
print("List without duplicates:", unique)


# 64. Reverse a list without using reverse()
lst = [1, 2, 3, 4, 5]
rev = lst[::-1]
print("Reversed list:", rev)


# 65. Sort list without using sort()
lst = [5, 2, 9, 1, 3]
for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] > lst[j]:
            lst[i], lst[j] = lst[j], lst[i]
print("Sorted list:", lst)


# 66. Count occurrences of element
lst = [1, 2, 2, 3, 2, 4]
element = 2
count = lst.count(element)
print("Occurrences of", element, ":", count)


# 67. Merge two lists
l1 = [1, 2, 3]
l2 = [4, 5, 6]
merged = l1 + l2
print("Merged list:", merged)


# 68. Find common elements in two lists
l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]
common = list(set(l1) & set(l2))
print("Common elements:", common)


# 69. Rotate list by k positions
lst = [1, 2, 3, 4, 5]
k = 2
rotated = lst[k:] + lst[:k]
print("Rotated list:", rotated)


# 70. Flatten a nested list
nested = [[1, 2], [3, 4], [5, 6]]
flat = [item for sublist in nested for item in sublist]
print("Flattened list:", flat)




# Module 6: Tuples, Sets, Dictionaries


# 71. Create a tuple and access elements
t = (10, 20, 30, 40)
print("Tuple:", t)
print("First element:", t[0])
print("Second element:", t[1])


# 72. Convert tuple to list
t = (1, 2, 3, 4)
lst = list(t)
print("Tuple:", t)
print("Converted List:", lst)


# 73. Count frequency of elements using dictionary
lst = [1, 2, 2, 3, 3, 3, 4]
freq = {}

for i in lst:
    freq[i] = freq.get(i, 0) + 1

print("Frequency:", freq)


# 74. Merge two dictionaries
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}

d1.update(d2)
print("Merged dictionary:", d1)


# 75. Find key with maximum value in dictionary
d = {'a': 10, 'b': 25, 'c': 15}
max_key = max(d, key=d.get)
print("Key with maximum value:", max_key)


# 76. Remove a key from dictionary
d = {'a': 1, 'b': 2, 'c': 3}
d.pop('b')
print("Dictionary after removing key:", d)


# 77. Create a set and perform union and intersection
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

print("Union:", s1.union(s2))
print("Intersection:", s1.intersection(s2))


# 78. Check if two sets are disjoint
s1 = {1, 2, 3}
s2 = {4, 5, 6}

print("Are sets disjoint?", s1.isdisjoint(s2))


# 79. Convert list into dictionary (index as key)
lst = ['a', 'b', 'c']
d = {i: lst[i] for i in range(len(lst))}
print("Dictionary:", d)


# 80. Sort dictionary by values
d = {'a': 3, 'b': 1, 'c': 2}
sorted_dict = dict(sorted(d.items(), key=lambda x: x[1]))
print("Sorted dictionary:", sorted_dict)



# Module 7: Functions

# 81. Write a function to check prime number
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Enter number: "))
print("Prime" if is_prime(num) else "Not Prime")


# 82. Function to calculate factorial using recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

print("Factorial:", factorial(5))


# 83. Function to check palindrome string
def is_palindrome(s):
    return s == s[::-1]

text = input("Enter string: ")
print("Palindrome" if is_palindrome(text) else "Not Palindrome")


# 84. Function to return sum of list elements
def list_sum(lst):
    return sum(lst)

numbers = [1, 2, 3, 4, 5]
print("Sum:", list_sum(numbers))


# 85. Function with default arguments
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Aryan")


# 86. Function using *args to sum numbers
def sum_numbers(*args):
    return sum(args)

print("Sum:", sum_numbers(1, 2, 3, 4))


# 87. Function using **kwargs to print key-value pairs
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

print_details(name="Aryan", age=21, city="Budaun")


# 88. Lambda function to find square of number
square = lambda x: x * x
print("Square:", square(5))


# 89. Higher-order function example
def apply_function(func, value):
    return func(value)

def double(x):
    return x * 2

print("Result:", apply_function(double, 5))


# 90. Write function to return Fibonacci sequence
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(10)




# Module 8: OOP Basics

# 91. Create a class Student with name and marks
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s1 = Student("Aryan", 85)
print("Name:", s1.name)
print("Marks:", s1.marks)


# 92. Add method to calculate grade
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        else:
            return "D"

s1 = Student("Aryan", 82)
print("Grade:", s1.calculate_grade())


# 93. Create a class Rectangle with area method
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

r = Rectangle(5, 3)
print("Area of Rectangle:", r.area())


# 94. Demonstrate inheritance with Animal → Dog
class Animal:
    def sound(self):
        print("Animals make sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.sound()
d.bark()


# 95. Implement encapsulation using private variable
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age   # private variable

    def get_age(self):
        return self.__age

p = Person("Aryan", 21)
print("Name:", p.name)
print("Age:", p.get_age())



# Module 9: File Handling & Exceptions

# 96. Write data to a file
with open("data.txt", "w") as file:
    file.write("Hello, this is a sample file.\n")
    file.write("Python file handling example.")
print("Data written to file.")


# 97. Read content from a file
with open("data.txt", "r") as file:
    content = file.read()
    print("File content:")
    print(content)


# 98. Count number of lines in a file
with open("data.txt", "r") as file:
    lines = file.readlines()
    print("Number of lines:", len(lines))


# 99. Handle division by zero using try-except
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")


# 100. Custom exception for invalid age
class InvalidAgeError(Exception):
    pass

try:
    age = int(input("Enter age: "))
    if age < 18:
        raise InvalidAgeError("Age must be 18 or above.")
    else:
        print("Valid age.")
except InvalidAgeError as e:
    print("Error:", e)
```
